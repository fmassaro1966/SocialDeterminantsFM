```python
plt.figure(figsize=(12,6))
plt.plot(
    df['M_NOVEH'],
    df['M_NOINT'],
    marker="+",
    linestyle="--",
    color="#37BA26",
    linewidth=0.5,
)

plt.xlabel("Households with no vehicle available (%)")
plt.ylabel("Households with no internet access (%)")
plt.title("What is the correlation with no vehicle and no internet access?", fontsize=18, pad=20)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.axvline(
    df['M_NOVEH'].mean(),
    color="#37BA26BA",
    linewidth=0.5,
    alpha=0.7)

ax.text(
    df['M_NOVEH'].mean() + 0.05,
    df['M_NOINT'].mean(),
    "Mean",
    color="#37BA26BA",
    fontsize=8
)
```
